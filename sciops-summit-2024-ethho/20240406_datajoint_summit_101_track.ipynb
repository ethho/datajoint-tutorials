## What is structured data and why is it important?

[According to Amazon Web Services (AWS)](https://aws.amazon.com/what-is/structured-data/):

> Structured data is data that has a standardized format for efficient access by software and humans alike. It is typically tabular with rows and columns that clearly define data attributes. Computers can effectively process structured data for insights due to its quantitative nature.

Formally, _structured data_ has:

1. **Definable Attributes**: We intuitively visualize these as the "column names" in a spreadsheet. Each entry has a value for each pre-defined attribute.
2. **Quantitative**: Structured data is usually quantitative. As scientists, we know first hand that quantitative data -- compared to qualitative data such as observations written in a lab notebook -- lets us make more powerful claims from our data, and more importantly, evaluate our error bounds. See **Appendix A** for an example of how 
3. **Relational Attributes**: Structured data often has common attributes that link related datasets. For example, a `students` dataset could contain an attribute that link each student entry to an entry in the `student_majors` dataset.

Though these properties may seem abstract, we interact with structured data in day-to-day life, whenever we use _spreadsheets_.
A typical Excel spreadsheet has a few pre-defined attributes (columns) that contain quantitative data.
We know from experience that storing and managing data in this structured manner allows us to analyze data quickly and easily; we `SUM()` up the total balance of our invoices, calculate the mean score on the test we just graded, etc.

Consider an example of a `students` dataset, stored in a spreadsheet:

| student_id | first_name | last_name | date_of_birth | sex | student_major | gpa |
|------------|------------|-----------|---------------|-----|---------------|-----|
| 001        | John       | Doe       | 2000-01-01    | M   | Computer Science | 3.5 |
| 002        | Jane       | Smith     | 2001-02-02    | F   | Mechanical Engineering | 4.0 |
| 003        | Jim        | Brown     | 2002-03-03    | M   | Mechanical Engineering | 3.8 |

Using spreadsheets, we can easily compute some derived values, such as the mean GPA, first name of the youngest student, or the ratio of M vs F vs Other students.
So spreadsheets are a first step in leveraging the power of structured data.



## Why aren't spreadsheets good enough?

Anyone who has done data analysis at scale knows that:

1. Spreadsheets are a good first step when you don't have much data (few entries), but
2. They don't scale well when your data scales to non-trivial (say, thousands of entries) scale, and
3. It's difficult to collaborate with others using spreadsheets. Yes, even using the collaborative features of Google Sheets.
4. Spreadsheets lack the features to define more complex structure in our data.

Let's go back to our `student` dataset example, and say for the sake of argument that we're a program coordinator for the College of Physical Sciences, seeking to increase enrollment.
At only 3 students and 1 employee, our dataset is easily managed in Excel.
Now imagine that our recruitment efforts are successful, and 2 years later:

- We have 2000 students enrolled, and
- To handle data entry for all the new students, we hire two assistant coordinators: Alice and Bob.

Even at this scale (which is still small relative to real scientific datasets), we start to encounter issues. For example:

1. When Bob creates new `student` entries, he uses different formats for the columns:
    - He uses date format `01-01-2000` instead of `2000-01-01` when he inserts `date_of_birth` for a new student.
    - He sometimes enters `sex` as `m` instead of `M`.
    - He often uses `Mech. Eng.` as shorthand for `Mechanical Engineering`.
2. `student_id` can only represent at most 1000 students as a 3 digit number! We need to go back and add 1-2 more leading digits to every `student_id` so that we can support 2000 students.
3. Alice and Bob have difficulty collaborating on the same dataset. To maximise efficiency, they decided to split up data entry for incoming students. Alice will take the first half, and add students to the top of the spreadsheet. Bob agrees to take the second half and append to the bottom. This is a perfectly reasonable strategy if you're using a spreadsheet, but it is error prone:
    - They sometimes create two new students with the same `student_id`.
    - When finals are done, they must recalculate and re-enter the `gpa` for each student. It's difficult to keep track of which students have already been updated. Or which students don't need an update because they dropped the course.

The core issue here is that we have a lot of structure in our data that spreadsheets can't handle.
For example, a spreadsheet can't force Bob to enter `M` instead of `m`.
It doesn't know that `student_id` should be unique, and can't enforce this as a contraint.
It can't catch errors when Alice enters a `student_major` that doesn't have a corresponding Department.

If we can represent and enforce this structure in the dataset, we have cleaner data, it's easier to maintain our data as it scales, and we can perform more advanced analysis on our data.

# Relational Databases to the Rescue

A tech savvy program coordinator might think: "I'll teach Alice and Bob Python! Then we can migrate the dataset to a CSV (for the record, a CSV file is effectively a spreadsheet), and my assistants can write Python functions that manage all of this for them."
But these issues are hard to fix with Python alone.
It would also take a lot of time for Alice and Bob to write this code.

Rather than write this software from scratch, we can leverage **relational databases**, defined [by AWS](https://aws.amazon.com/relational-database/) as:

> A collection of data points with pre-defined relationships between them.
>
> You can join the tables in a relational database to gain deeper insights into the interconnection between diverse data points. For example, the table `customer` has columns `name` and `address` with row values `John` and `California`. Similarly, the table `product` has columns `name` and `cost` with row values `shoes` and `$12`. You join the `customer` table with the `product` table to identify popular products that customers purchase more frequently.

In addition, a database can be configured to automatically enforce assumptions in our data, such as:

- Only allow `sex` to be `M`, `F`, or `U` (undefined), not `m` or any other value.
- Only allow `student_major` to be the name of a Department in the College.
- Ensure that there are no students with the same `student_id`. In other words, `student_id` is **unique**.

Databases also empower Alice and Bob to collaborate efficiently. Using (default settings of) a database, they cannot edit `gpa` for the same student simultaneously.
If there is uncertaintly about whether `gpa` was edited twice, the database maintains an auditable log of changes, and we can roll back to a previous snapshot of the database to resolve the conflict. This approach to collaboration scales well, even when we expand the team to 100 data analysts.

# Relational Databases Are Hard to Use

Despite their power to represent complex relationships in our data, relational databases (RDBMS) have a steep learning curve for both users and administrators.
For example, here is a code snippet in Structured Query Language (SQL) that creates tables for Students, StudentMajors, and Departments for us:

```sql
CREATE TABLE Student (
    student_id INT UNSIGNED,
    first_name VARCHAR(40),
    last_name VARCHAR(40),
    sex ENUM('F', 'M', 'O'),
    date_of_birth DATE,
    home_address VARCHAR(120),
    home_city VARCHAR(60),
    home_state CHAR(2),
    home_zip CHAR(10),
    home_phone VARCHAR(29),
    PRIMARY KEY (student_id)
);

CREATE TABLE Department (
    dept VARCHAR(6),
    dept_name VARCHAR(200),
    dept_address VARCHAR(200),
    dept_phone VARCHAR(20),
    PRIMARY KEY (dept)
);

CREATE TABLE StudentMajor (
    student_id INT UNSIGNED,
    dept VARCHAR(6),
    declare_date DATE,
    PRIMARY KEY (student_id, dept),
    FOREIGN KEY (student_id) REFERENCES Student(student_id),
    FOREIGN KEY (dept) REFERENCES Department(dept)
);
```

It's not important to know what this code is doing; the point here is just to convey that setting up a database using SQL is not trivial.

For users (such as our 2 data analysts), SQL is also less user-friendly than Excel formulae.
Instead of `=MIN(G:G)` to find the oldest of all students, we'd write the following in SQL:

```sql
SELECT CONCAT(first_name, ' ', last_name) AS full_name
FROM Student
ORDER BY date_of_birth
LIMIT 1;
```

While these SQL examples aren't terribly complex, they become much more complex, and therefore unapproachable to beginners, as the data model expands.
In addition, teams using databases need to host the database somewhere, either by running the database software on their local machine (which makes collaboration difficult), or deploying it on a commercial cloud provider such as Amazon Web Services (AWS) or Google Cloud (which requires even more IT knowledge to set up and administer). 

# DataJoint: Making Relational Databases Easy

Between the steep learning curve for SQL and added complexity of deploying a database, most research labs shy away from using relational databases.
But recall that relational databases have a lot to offer in terms of modeling and managing scientific data, especially since these datasets tend to be highly interconnected and complex.

The original author of DataJoint (Dimitri Yatsenko) recognized this tradeoff, and sought to develop a tool that lowered the barrier of entry for using relational databases.
The product of this effort is the [`datajoint` library](https://datajoint.com/docs/core/datajoint-python/0.14/concepts/principles/#theoretical-foundations), which you can pip install:

````python
!pip install -q datajoint
````

In short, `datajoint` empowers teams to leverage relational databases to represent complex structure in their data, but without the added complexity of administering and querying databases using SQL. In addition, `datajoint` makes it easy to build _computational data pipelines_ using structured data; this is why we often refer to `datajoint` as a system for managing a **computational database**.

These benefits are easier to show than to tell, however, so let's return to our `Students` dataset example.
We first import the `datajoint` package and define a **schema** that will contain many **tables** such as `Student`, `StudentMajor`, `Department`, etc.

````python
import datajoint as dj

# TODO: remove
dj.config['database.host'] = 'fakeservices.datajoint.io'
dj.config['database.user'] = 'root'
dj.config['database.password'] = 'password'
dj.config['database.use_tls'] = False

schema = dj.Schema('djtest.university')
````

We can name the `schema` anything, though it _should_ be unique in your database.
We can now define our first table: `Student`.

````python
@schema
class Student(dj.Manual):
    definition = """
    student_id : int unsigned   # university-wide ID number
    ---
    first_name      : varchar(40)
    last_name       : varchar(40)
    sex             : enum('F', 'M', 'U')
    date_of_birth   : date
    home_address    : varchar(120) # mailing street address
    home_city       : varchar(60)  # mailing address
    home_state      : char(2)      # US state acronym: e.g. OH
    home_zip        : char(10)     # zipcode e.g. 93979-4979
    home_phone      : varchar(29)  # e.g. 414.657.6883x0881
    """
````

There's a lot to unpack here, so let's walk through step-by-step:

1. We declare a table using a custom Python `class` of the same name.
    - This `class` inherits from `datajoint.Manual`. It's fine if you don't know what this means: just note that `dj.Manual` denotes this as a table where we will manually (as opposed to automatically via a pipeline) enter data.
    - We include this new table in the `schema` we just created using the decorator syntax `@schema`. Again, you don't need to know how decorators work to use `datajoint`. Just include the `@schema` line above any table that should be in the `schema`.
2. We declare a single `definition` attribute in this class. This tells `datajoint` which attributes to set up and what their types are. Each attribute should be on its own line, with the name on the left of the `:` and the type and description on the right.
    - `student_id` is an integer (`int unsigned`) that identifies a single student in this table.
    - `first_name` and `last_name` can be strings (`varchar`) of maximum 40 characters in length.
    - We allow `sex` to be only one of `F`, `M`, or `U`.
    - `date_of_birth` is represented unambiguously as a `date` type.
    - `home_state` has type `char(2)` instead of `varchar`. This means that the value must be _exactly_ 2 characters, instead of `varchar` which would allow 1 or 2 (or 0).
3. Finally, we denote [**primary key attributes**](https://www.ibm.com/docs/en/ida/9.1.1?topic=entities-primary-foreign-keys) by defining them above the triple dash `---` separator.
    - **Primary key attributes**, by definition, uniquely identify this entry in the table.
    - Note that this helps us solve the "duplicate `student_id` problems" that Alice and Bob were having!

Similarly, we can declare another table that stores `Department` entries.
Like `Student`, we declare it as a `dj.Manual` table, meaning that this table will be empty until we insert entries manually.

````python
@schema
class Department(dj.Manual):
    definition = """
    dept : varchar(6)   # abbreviated department name, e.g. BIOL
    ---
    dept_name    : varchar(200)  # full department name
    dept_address : varchar(200)  # mailing address
    dept_phone   : varchar(20)
    """
````

Like `Student`, we define a primary key, but this time it's the `dept` attribute of type `varchar` (maximum length of 6).
We also declare attributes for the full name and contact info for the `Department`.

So far, there is no association between `Student` and `Department` tables.
We could insert entries (rows) into each of these tables and use them independently, and that's a completely valid use of `datajoint` and databases.
But the real power appears when we start to represent and use relationships between tables.

In this example, we'll associate `Student` and `Department` by declaring a third table:

````python
@schema
class StudentMajor(dj.Manual):
    definition = """
    -> Student
    ---
    -> Department
    declare_date :  date  # when student declared her major
    """
````

# Appendix A: A story about structured data

Consider the following excerpt from my lab notebook, from my days as an experimental biochemist.
It describes the results of a cleaving experiment, where a sample of our protein of mass 50 kDa was incubated with (experimental group) and without (control) TeV protease. We separated the resulting products on a gel and imaged it:

```
A strongly visible band was observed at about 50 kilodaltons (kDa) for the control group.
A weakly visible band was observed at about 50 kDa for the sample containing TeV protease.
```

As human researchers with domain expertise, we can make some intuitive conclusions from these observations (for example, I would say that TeV protease cleaves our protein of interest with moderate efficiency).
With only qualitative data, however, our analytical toolbox was limited, and we could only make weak claims about how our system is behaving.

At some point, our group wanted to answer some harder questions about this system.
We could tell from this preliminary experiment that TeV protease is cleaving, but we now want to know:

- How much of our protein is TeV cleaving?
- How much TeV is required for cleavage to occur?
- How quickly is this happening?

We needed **structured data** to answer these questions, so we additionally started collecting quantitative metrics from our gel images.
This allowed us to quantify the exact size in kDa (y-axis) and intensity of each band in the gel (Figure 1).
In addition, we could quantify the increase in cleavage efficiency as we increased the concentration of TeV (from left to right).
More importantly, this allowed us to quantify the uncertainty in our experiments, by calculating the variance and error bounds of these metrics.

````python
@schema
class Course(dj.Manual):
    definition = """
    -> Department
    course  : int unsigned   # course number, e.g. 1010
    ---
    course_name :  varchar(200)  # e.g. "Neurobiology of Sensation and Movement."
    credits     :  decimal(3,1)  # number of credits earned by completing the course
    """
    
@schema
class Term(dj.Manual):
    definition = """
    term_year : year
    term      : enum('Spring', 'Summer', 'Fall')
    """

@schema
class Section(dj.Manual):
    definition = """
    -> Course
    -> Term
    section : char(1)
    ---
    auditorium   :  varchar(12)
    """
    
@schema
class CurrentTerm(dj.Manual):
    definition = """
    -> Term
    """

@schema
class Enroll(dj.Manual):
    definition = """
    -> Student
    -> Section
    """

@schema
class LetterGrade(dj.Manual):
    definition = """
    grade : char(2)
    ---
    points : decimal(3,2)
    """

@schema
class Grade(dj.Manual):
    definition = """
    -> Enroll
    ---
    -> LetterGrade
    """
````

````python
# Enrollment not matching major 
Enroll - StudentMajor 
````

````python
# Total grade points
(Course * Grade * LetterGrade).proj(total='points*credits')
````

## Aggr

````python
# Students in each section
Section.aggr(Enroll, n='count(*)')
````

````python
# Average grade in each course
Course.aggr(Grade*LetterGrade, avg_grade='avg(points)')
````

````python
# Fraction graded in each section
(Section.aggr(Enroll,n='count(*)') * Section.aggr(Grade, m='count(*)')).proj(
    'm','n',frac='m/n')
````

````python
# Student GPA
Student.aggr(Course * Grade * LetterGrade, 
             gpa='sum(points*credits)/sum(credits)')
````

````python
# Average GPA for each major
gpa = Student.aggr(Course * Grade * LetterGrade,
                   gpa='sum(points*credits)/sum(credits)')
Department.aggr(StudentMajor*gpa, avg_gpa='avg(gpa)')
````
